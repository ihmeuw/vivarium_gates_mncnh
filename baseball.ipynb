# Baseball Stats Notebook

A simple notebook to test papermill parameterization via vivarium cluster tools.

```python
# Parameters (injected by papermill)
player_name = None
stat_category = None
```

```python
PLAYER_STATS = {
    "shohei_ohtani": {
        "batting_average": ".274",
        "home_runs": "225",
        "rbi": "600",
        "stolen_bases": "112",
        "ops": ".942",
    },
    "cal_raleigh": {
        "batting_average": ".232",
        "home_runs": "128",
        "rbi": "310",
        "stolen_bases": "5",
        "ops": ".800",
    },
}
```

```python
player = player_name.lower()
category = stat_category.lower()

if player not in PLAYER_STATS:
    print(f"Unknown player: {player_name}")
    print(f"Available players: {', '.join(PLAYER_STATS.keys())}")
else:
    stats = PLAYER_STATS[player]
    if category not in stats:
        print(f"Unknown stat category: {stat_category}")
        print(f"Available categories: {', '.join(stats.keys())}")
    else:
        print(f"{player_name} - {stat_category}: {stats[category]}")
```
